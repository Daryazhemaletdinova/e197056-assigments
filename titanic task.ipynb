```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_2_train_file = '/Users/darya/Desktop/titanic.csv'
df = pd.read_csv(path_2_train_file)
print(df)
```

```python
df.head(20)
```

```python
#Question1. How many people on the titanic?
df["PassengerId"]
```

```python
#How many survived?
len(df[df["Survived"]==1])
```

```python
#How many is not survived?
len(df[df["Survived"]==0])
```

```python
#How many is not survived and how many survived?
print(df["Survived"] .value_counts())
```

```python
#How many that survived were female and how many that died were female?
print(df["Survived"] [df["Sex"] =='female'].value_counts())
```

```python
#Question 3 & 4 How many children were on the titanic, NB: you are a child if age < 17.; How many died? 

print(df["Survived"] [df["Age"] <17].value_counts())
```

```python
#Question 5: How many people had families with them?
df["Family"] =  df["Parch"] + df["SibSp"]
df["Family"].loc[df['Family'] > 0] = 1
df["Family"].loc[df['Family'] == 0] = 0
print(df["Family"].value_counts())
```

```python
#Question 6: What is the ratio of female to male?

df.groupby(['Survived','Sex']).count().Name
```

```python
df[["Sex", "Survived"]].groupby(["Sex"], as_index=False).mean().sort_values(by="Survived",ascending=False)
```

```python
#Question 7: What contributed to the survival of those who survived?
#Age feature check
df[["Sex", "Survived"]].groupby(["Sex"], as_index=False).mean().sort_values(by="Survived",ascending=False)
```

```python
#Pclass check
df[["Pclass", "Survived"]].groupby(["Pclass"], as_index=False).mean().sort_values(by="Survived",ascending=False)
```

```python
#Number of siblings
df[["SibSp", "Survived"]].groupby(["SibSp"], as_index=False).mean().sort_values(by="Survived",ascending=False)
```

```python
#Number of parents
df[["Parch", "Survived"]].groupby(["Parch"], as_index=False).mean().sort_values(by="Survived",ascending=False)
```
